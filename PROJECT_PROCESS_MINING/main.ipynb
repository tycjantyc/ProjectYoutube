```python
download = True
limit = 200

if download:
    for video_url in video_urls:
        stats = get_video_details(video_url, API_KEY)

        if int(stats['views']) < view_count_threshold and stats['duration'].total_seconds() < max_duration:
            download_video(video_url)
            donwloaded_num += 1
        
        if donwloaded_num == limit:
            break
        
    print(f"Downloads {donwloaded_num} videos")
```
